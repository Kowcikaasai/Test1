```python
from bs4 import BeautifulSoup
```

```python
str="""<html>
<head>
<title>A simple example page</title>
</head>
<body>
<p>Here is some simple content for this page.</p>
<div class="city">
<h2>London</h2>
<p>London is the capital of England.</p>
</div>
<div class="country">
<h2>UK and England</h2>
<p>England is one country, and the United Kingdom is four countries united together.</p>
</div>
</body>
</html>"""
soup=BeautifulSoup(str,"html.parser")
print(soup)
print(soup.prettify())
```

```python
from bs4 import BeautifulSoup
import requests
url = 'https://ravindrababuravula.com/courses.php'
reqs = requests.get(url)
soup=BeautifulSoup(reqs.text,"html.parser")
print(soup)
print(soup.prettify())
```

```python
soup.find("div",class_="edugate-content").a.text
```

```python
c=[]
for b in soup.find_all("div", class_="edugate-content"):
  print(b.a.text)
  c.append(b.a.text)
c
```

```python
d=[]
for b in soup.find_all("div", class_="description"):
  print(b.text, "\n")
  d.append(b.text)
d
```

```python
import pandas as pd
e = pd.DataFrame(list(zip(c,d)))
e
```
